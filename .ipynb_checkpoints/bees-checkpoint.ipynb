Next we'll import the libraries we will need.

```python
import tensorflow as tf
import tensorflow_datasets as tfds
import matplotlib.pyplot as plt
from keras import layers
```

### Loading the Dataset
Now we can load the dataset. We will put 80% of the dataset into a training set, and the remaining 20% into a test set.

The dataset contains images of bees and wasps, each image has four corresponding labels indicating the presence of different characteristics.

```python
bee_train, info = tfds.load(
        'bee_dataset/bee_dataset_150',
        as_supervised=True,
        split='train[:80%]',
        with_info=True
        )
bee_test = tfds.load(
        'bee_dataset/bee_dataset_150',
        as_supervised=True,
        split='train[80%:]',
        )
```

Let's take a look at some of the images in the training set.

```python
# Display samples from the dataset with labels [cooling, pollen, varroa, wasp]
def show_samples(ds):
    ds.shuffle(64)
    plt.figure(figsize=(12, 12))
    for i, (image, labels) in enumerate(ds.take(9)):
        plt.subplot(3, 3, i+1)
        plt.imshow(image.numpy().astype("uint8"))
        plt.title([
            labels['cooling_output'].numpy(),
            labels['pollen_output'].numpy(),
            labels['varroa_output'].numpy(),
            labels['wasps_output'].numpy()
        ])
        plt.axis("off")
```

Labels are cooling, pollen, varroa, and wasp from left to right.

(Note: You can rerun the cell below to get new samples)

```python
show_samples(bee_train)
```

### Data Augmentation
We see that the dataset contains images of bees with the head of the bee either at the top or bottom of the image. In the data collected from the real world it is unlikely the bees would be so nicely organized. We can make our dataset more representative of real world data using data augmentation. Training the model on unaugmented data would likely result in the model performing better on the dataset, but the model would not perform to expectations when introduced to new data.

We will introduce a random rotation and random zoom. We also add padding to the images to make them square, to avoid losing as much data in the rotations. The augmentations we select are representative of what we would expect to see in real world data. The bees would have a random orientation relative to the camera, and there might be slight differences in distance to the camera, which we represent with a slight random zoom.

We could integrate the augmentations into the layers of our model, making the code cleaner. We leave it seperate for demonstration purposes, so we can observe the results.

```python
pad = layers.ZeroPadding2D(padding=((0,0),(37,38)))
# The pad layer expects a batched dataset, we then unbatch it to access the samples
bee_train = bee_train.batch(64).map(lambda x, y: (pad(x), y)).unbatch()
bee_test = bee_test.batch(64).map(lambda x, y: (pad(x), y)).unbatch()

augmentation = tf.keras.Sequential([
    layers.RandomRotation(0.4, fill_mode='constant'),
    layers.RandomZoom(0.1, fill_mode='constant')
])

bee_train = bee_train.map(lambda x, y: (augmentation(x), y))
```

We can look at the results of our augmentation.

(Note: You can rerun the cell below to get new samples)

```python
show_samples(bee_train)
```

### Assessing the Data
Now let's examine the distribution of our target characteristics in the datasets.

```python
def plot_distribution(ds):
    labels = ds.map(lambda x, y: y)
    df = tfds.as_dataframe(labels).sum()
    df.plot.bar()
    plt.show()

print('Training Set Distribution (5992 total items)')
plot_distribution(bee_train)
print('Test Set Distribution (1498 total items)')
plot_distribution(bee_test)
```

We should consider a few things when examining the distribution of the datasets.

It is important that there aren't any significant discrepancies between the training and test set. We can see that the test set has slightly lower occurence of wasps in it, but it should be acceptable for our purposes.

We can also see that for each characteristic, it is much more likely for the characteristic to be absent than present. This may cause the model to skew towards predicting the absence of characteristics. Outside of collecting more data, there are two ways we might deal with this issue. The first is downsampling, eliminating items from the dataset to reduce the difference. Our dataset is too small to comfortably downsample. The second option would be upsampling, creating copies of the elements of the dataset that we wish to increase the occurence of. When upsampling, we could also use data augmentation, so that the copies we make are not exact duplicates.

For our purposes, however, these issues with the dataset won't cause significant problems. We will keep them in mind while we analyze the models performance, though.

### Designing the Model
Now we can design our model. We can consider each of our target characteristics to be a binary classification problem. We will define our model as a subclass of the Keras Model class. This will enable us to easily create and train models for each of our targets.

```python
class BeeModel(tf.keras.Model):
    LOSS = tf.keras.losses.BinaryCrossentropy(from_logits=False)
    METRICS = [
        tf.keras.metrics.BinaryAccuracy(),
        tf.keras.metrics.Precision(),
        tf.keras.metrics.Recall()
        ]
    
    def __init__(self):
        super(BeeModel, self).__init__()
        self.flatten = layers.Flatten()
        self.normalize = layers.Rescaling(1./255)
        self.conv_1 = layers.Conv2D(16, 3, activation='relu', input_shape=(150, 150, 3))
        self.conv32 = layers.Conv2D(32, 3, activation='relu')
        self.maxpool = layers.MaxPooling2D()
        self.d128 = layers.Dense(128, activation='relu')
        self.d64 = layers.Dense(64, activation='relu')
        self.d1 = layers.Dense(1, activation='sigmoid')
        self.compile(optimizer='adam', loss=self.LOSS, metrics= self.METRICS)

    def call(self, x):
        x = self.normalize(x)
        x = self.conv_1(x)
        x = self.maxpool(x)
        x = self.conv32(x)
        x = self.maxpool(x)
        x = self.flatten(x)
        x = self.d128(x)
        x = self.d64(x)
        return self.d1(x)
```

We define the layers that we will use in our model, and compile the model with the necessary optimizer, loss function, and metrics. We then arrange the layers into a series of steps to be used in the call function.

Let's discuss the layers in more detail:

* **Rescaling:** This is a preprocessing layer that will normalize the pixel values to a float between 0 and 1.
* **Connv2d:** Convolutional layers enable the neural network to efficiently process image data. Each node in a convolutional layer is activated by a small sector of adjacent values in the previous layer. For the first convolutional layer this is a 3x3 area of pixels in the image. Convolutional layers allow the model to detect patterns in the data that are spatially near in the image. Since patterns between pixels in close proximity are much more likely to be relevant than patterns in disparate pixels, convolutional layers efficiently extract visual features from the image.
* **MaxPooling2d:** The maxpooling layer reduces the number of nodes in the network, it connects to spatial window in the previous layer and simply takes on the maximum value of the connected nodes. 
* **Flatten:** This reduces the dimensions of the data, converting the multi-dimensional data into a single dimension.
* **Dense:** Each node in a dense layer is activated by all of the nodes in the previous layer. The dense layers in this model convert the features extracted by the convolutional layers down to a single binary value, that will predicts the presence of abscence of the target characteristic.



### Training
Now with our model class defined we can make and train a model for a particular characteristic.

```python
# Constants
BATCH = 64
EPOCHS = 12

# Make and train models
c_train = bee_train.map(lambda x, y: (x, y['cooling_output'])).batch(BATCH)
c_test = bee_test.map(lambda x, y: (x, y['cooling_output'])).batch(BATCH)
c_model = BeeModel()
c_log = c_model.fit(c_train, validation_data=c_test, epochs=EPOCHS)
```

```python
p_train = bee_train.map(lambda x, y: (x, y['pollen_output'])).batch(BATCH)
p_test = bee_test.map(lambda x, y: (x, y['pollen_output'])).batch(BATCH)
p_model = BeeModel()
p_log = p_model.fit(p_train, validation_data=p_test, epochs=EPOCHS)
```

```python
v_train = bee_train.map(lambda x, y: (x, y['varroa_output'])).batch(BATCH)
v_test = bee_test.map(lambda x, y: (x, y['varroa_output'])).batch(BATCH)
v_model = BeeModel()
v_log = v_model.fit(v_train, validation_data=v_test, epochs=EPOCHS)
```

```python
w_train = bee_train.map(lambda x, y: (x, y['wasps_output'])).batch(BATCH)
w_test = bee_test.map(lambda x, y: (x, y['wasps_output'])).batch(BATCH)
w_model = BeeModel()
w_log = w_model.fit(w_train, validation_data=w_test, epochs=EPOCHS)
```

### Examining Results
Now that we have our models trained, we will examine the metrics from the training history. We can access this data through the history object returned by the fit method.

```python
# Plot model training history
def plot_history(log, name='Model'):
    plt.figure(figsize=(12, 12))
    plt.subplot(2, 2, 1)
    plt.plot(range(log.params['epochs']), log.history['binary_accuracy'], label='Training Accuracy')
    plt.plot(range(log.params['epochs']), log.history['val_binary_accuracy'], label='Validation Accuracy')
    plt.legend(loc='lower right')
    plt.title('{} Accuracy'.format(name))

    plt.subplot(2, 2, 2)
    plt.plot(range(log.params['epochs']), log.history['loss'], label='Training Loss')
    plt.plot(range(log.params['epochs']), log.history['val_loss'], label='Validation Loss')
    plt.legend(loc='upper right')
    plt.title('{} Loss'.format(name))

    plt.subplot(2, 2, 3)
    plt.plot(range(log.params['epochs']), log.history['precision'], label='Training Precision')
    plt.plot(range(log.params['epochs']), log.history['val_precision'], label='Validation Precision')
    plt.legend(loc='lower right')
    plt.title('{} Precision'.format(name))

    plt.subplot(2, 2, 4)
    plt.plot(range(log.params['epochs']), log.history['recall'], label='Training Recall')
    plt.plot(range(log.params['epochs']), log.history['val_recall'], label='Validation Recall')
    plt.legend(loc='lower right')
    plt.title('{} Recall'.format(name))

    plt.show()


plot_history(c_log, name='Cooling Model')
```

Above we can see the results of training the model on the cooling output. We are tracking the following metrics during training:
* **Accuracy:** The ratio of correct predictions to total predictions of the model.
* **Loss:** The loss as computed by the loss function, in this case binary cross entropy.
* **Precision:** The ratio of correct positives predicted to the total number of predicted positives. In other words, how likely it is for an identified characteristic to be correct.
* **Recall:** The ratio of correct positive predictions to the total positive occurences in the dataset. That is to say, how likely it is that a characteristic will be detected by the model.

As we discussed earlier when examining the dataset, the dataset we are using is biased. There are significanlty more negative characteristics than positve. This means we cannot rely on accuracy alone. To demonstrate this, imagine a model that only predicted the absence of any characteristics. Such a model would acheive an accuracy greater than 80%, while being completely useless for any practical application. For this reason, we include precision and recall, so that we can more accurately assess how well our model is performing on positive predictions. Recall is of particular importance, as it measures how sensitive the model is to detecting characteristics, and is the most impacted by the bias in our dataset.

It is important to look for large differences between the training and validation metrics, as this will indicate overfitting. Overfitting occurs when the model is fit too strongly to the training set, making predictions based on patterns in the training set that are not true indicators of the target. There are many ways to combat overfitting. Reducing the training epochs, reducing the capacity of the neural network, or introducing a dropout layer will all help to mitigate overfitting. The data augmentation we applied earlier also contributes to reducing overfitting.

Below we will show some example predictions. The true values are above the image, with the predictions below.

```python
def get_predictions(img):
    img = tf.keras.utils.img_to_array(img)
    # The model expects batched images, so we expand the array
    img = tf.expand_dims(img, 0)
    c_pred = c_model.predict(img)
    p_pred = p_model.predict(img)
    v_pred = v_model.predict(img)
    w_pred = w_model.predict(img)
    predictions = [c_pred[0][0], p_pred[0][0], v_pred[0][0], w_pred[0][0]]
    return predictions


def show_sample_pred(ds):
    ds = ds.shuffle(64)
    plt.figure(figsize=(12, 12))
    for i, (image, labels) in enumerate(ds.take(9)):
        plt.subplot(3, 3, i+1)
        plt.imshow(image.numpy().astype("uint8"))
        plt.title([
            labels['cooling_output'].numpy(),
            labels['pollen_output'].numpy(),
            labels['varroa_output'].numpy(),
            labels['wasps_output'].numpy()
        ])
        plt.yticks([])
        plt.xticks([])
        plt.xlabel([ '%.3f' % pred for pred in get_predictions(image)])
```

Labels are cooling, pollen, varroa, and wasp from left to right.

### Conclusion
In this notebook, we implemented a neural network for the classification of bee images. We were able to acheive good performance on a variety of target characteristics by defining our models structure in a subclass of the keras Model class. If you are interested in how we might improve the model further there are a couple of options that would make sense to investigate:
* **Optimize the Design:** The models for the individual characteristics are structured similarly, but each model is trained in isolation. This is not very efficient. It might make sense to have the images first processed by a universal block designed to process the images into a usable form and then have specialized heads that process the output of that model to identify specific characteristics.
* **Transfer Learning:** We intentionally built our model from the ground up for demonstration purposes, but we would likely acheive better performance with less training by building our mobel on top of a pretrained model designed to extract visual features. Many pretrained models are publicly available. We could look for one that would be suitable for processing the bee images, and design our model to predict the target characteristics from the output of the pretrained model.
